```python
##Step 2: Data Cleaning and Preprocessing
##Clean the data: Handle missing values, convert dates, create a unified "salary" column (average hourly rate for hourly jobs, budget for fixed), 
##extract keywords from titles, and identify remote jobs
```

```python
# Convert published_date to datetime
df['published_date'] = pd.to_datetime(df['published_date'])
```

```python
columns_to_fill = ['country', 'title', 'link', 'budget', 'is_hourly', 'hourly_high', 'hourly_low', 'hourly_avg']

for col in columns_to_fill:
    if col in df.columns:
        if df[col].dtype == 'O':  # For object/string columns
            df[col] = df[col].fillna('Unknown')
        else:
            df[col] = df[col].fillna(0)
```

```python
# Handle salary: Create a 'salary' column
df['hourly_avg'] = (df['hourly_low'] + df['hourly_high']) / 2
df['salary'] = np.where(df['is_hourly'], df['hourly_avg'], df['budget'])
```

```python
# Fill missing salaries with median (separate for hourly/fixed)
median_hourly = df[df['is_hourly']]['salary'].median()
median_fixed = df[~df['is_hourly']]['salary'].median()
df.loc[df['is_hourly'] & df['salary'].isna(), 'salary'] = median_hourly
df.loc[~df['is_hourly'] & df['salary'].isna(), 'salary'] = median_fixed
```

```python
# Extract keywords from title
stop_words = set(stopwords.words('english'))
def extract_keywords(title):
    title = re.sub(r'[^a-zA-Z\s]', '', title.lower())  # Clean text
    tokens = word_tokenize(title)
    keywords = [word for word in tokens if word not in stop_words and len(word) > 2]
    return ' '.join(keywords)

df['keywords'] = df['title'].apply(extract_keywords)
```

```python
df.isnull().sum()
```

```python
# Identify remote jobs (simple keyword search)
df['is_remote'] = df['title'].str.lower().str.contains('remote')
```

```python
# Add month-year for time-based analysis
df['month_year'] = df['published_date'].dt.to_period('M')
```

```python
# Save cleaned data
df.to_csv('cleaned_jobs.csv', index=False)
print("Data cleaned and saved.")
```

```python
##Step 3: Task 1 - Analyze Correlation Between Job Title Keywords and Offered Salaries
##Use TF-IDF to vectorize keywords and linear regression to find correlations with salaries. Visualize with a heatmap or scatter plot.
```

```python
# Reload cleaned data if needed
df1 = pd.read_csv('cleaned_jobs.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Fill NaN with empty string
df1['keywords'] = df1['keywords'].fillna('')
# Vectorize keywords
tfidf = TfidfVectorizer(max_features=50)  # Top 50 keywords
X = tfidf.fit_transform(df1['keywords'])
feature_names = tfidf.get_feature_names_out()
```

```python
# Prepare data for regression
y = df1['salary']
X_dense = X.toarray()
X_train, X_test, y_train, y_test = train_test_split(X_dense, y, test_size=0.2, random_state=42)

# Train linear regression
model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error: {mse}")
```

```python
# Get correlations (coefficients)
correlations = pd.DataFrame({'Keyword': feature_names, 'Coefficient': model.coef_})
correlations = correlations.sort_values(by='Coefficient', ascending=False)
print(correlations.head(10))  # Top correlated keywords
```

```python
# Visualization: Bar plot of top correlations
plt.figure(figsize=(10, 6))
sns.barplot(x='Coefficient', y='Keyword', data=correlations.head(10))
plt.title('Top Keywords Correlated with Salaries')
plt.show()
```

```python
##Step 7: Task 5 - Create a Job Recommendation Engine Based on Current Job Postings
##Build a simple content-based recommender using TF-IDF and cosine similarity. Expose via Flask API.
```

```python
# Vectorize titles
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df1['title'])
```

```python
# Function for recommendations
def recommend_jobs(user_query, top_n=5):
    query_vec = vectorizer.transform([user_query])
    sim_scores = cosine_similarity(query_vec, tfidf_matrix).flatten()
    top_indices = sim_scores.argsort()[-top_n:][::-1]
    return df1.iloc[top_indices][['title', 'link', 'salary', 'country']]
```

```python
# Example usage
print(recommend_jobs('python developer'))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Prepare data
data = market_trend.rename(columns={'ds':'ds','y':'y'})

# Ensure 'ds' is datetime
data['ds'] = pd.to_datetime(data['ds'])

# Sort by date
data = data.sort_values('ds').reset_index(drop=True)

# Create numeric time index
data['t'] = np.arange(len(data))

# Fit linear trend
coeffs = np.polyfit(data['t'], data['y'], 1)

# Forecast next 6 months
forecast_t = np.arange(len(data), len(data)+6)
forecast_base = np.polyval(coeffs, forecast_t)

# Scenarios
optimistic = forecast_base * 1.1
pessimistic = forecast_base * 0.9

# Future dates
future_dates = pd.date_range(start=data['ds'].iloc[-1], periods=7, freq='M')[1:]

# Plot
plt.figure(figsize=(10,6))
plt.plot(data['ds'], data['y'], label='Actual', marker='o')
plt.plot(future_dates, forecast_base, label='Base Forecast', marker='o')
plt.plot(future_dates, optimistic, label='Optimistic (+10%)', marker='o')
plt.plot(future_dates, pessimistic, label='Pessimistic (-10%)', marker='o')
plt.title('Remote Work Forecast - Scenarios')
plt.xlabel('Month')
plt.ylabel('Remote Work %')
plt.legend()
plt.show()
```
